# Project: Web Scraping from an E-Book Store

```python
# importing all the useful libraries
from bs4 import BeautifulSoup       # for web scraping
import requests                     # for making HTTP requests
import pandas as pd                 # for creating & managing dataframe
```

```python
Book_Title = []                     # for storing Book Titles
Price = []                          # for storing Book Prices
Stock_Availibility = []             # for storing whether book is available or not
Product_Link = []
```

```python
domain = 'https://books.toscrape.com'     # Domain of the website 'Books to scrape'
```

```python
for pg in range(1,51):     
    url = 'https://books.toscrape.com/catalogue/page-'+str(pg)+'.html' 
    #print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    #print(soup)

    h3 = soup.find_all('h3')        # for finding titles of all the books
    #print(h3)

    for i in h3:
        book = i.find('a')
        Book_Title.append(book.text)
        link = book.get('href')     # for finding url of product
        #print(link)
        link = domain + '/catalogue' + '/' + link
        Product_Link.append(link)

    p_price = soup.find_all('p', class_ = 'price_color')        # for finding prices of all the books
    #print(p_price)

    for price in p_price:
        book_price = price.text
        book_price = book_price.replace('Ã‚', '')
        Price.append(book_price)

    p_stock = soup.find_all('p', class_ = "instock availability")       # for finding stock availability of all the books
    
    for stock in p_stock:
        book_stock = stock.text.strip()
        Stock_Availibility.append(book_stock)

    print(f'Page {pg} data has been scrapped successfully!')
```
